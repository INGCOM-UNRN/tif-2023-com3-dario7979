## Calculadora Clásica

Primero Solicita al usuario que ingrese la operación que desea realizar, por ejemplo: "2 + 3 - 4 : 5 * 6". Los elementos de la operación deben estar separados por espacios.

Luego, el programa separa los elementos de la operación en una lista llamada elementos.

Recorre cada elemento de la lista "elementos" y realiza las operaciones correspondientes. 

Finalmente, se solicita al usuario que ingrese "=" para obtener el resultado. 

```python
def calculadora_clasica():
    """ 
    Realiza operaciones de suma, resta, multiplicación y división de numeros enteros y flotantes.
    El programa se ejecuta hasta que el usuario elige la opción "4" para salir.
    """
    
    print("¡Bienvenido a la calculadora clásica!")
    operacion = input("Ingrese la operación que desea realizar (ejemplo: 2 + 3 - 4 : 5 * 6 ) : ")
    operacion = operacion.replace(" ", "")  # Elimina los espacios en blanco de la operación ingresada

    elementos = []  # Lista para almacenar los elementos de la operación
    numero_actual = ""  # Variable para construir el numero actual.
    for caracter in operacion:
        if caracter.isdigit() or caracter == ".":
            numero_actual = numero_actual + caracter   # Agrega el caracter a la construcción del número actual si es un dígito o un punto decimal
        else:
            if numero_actual: # Verifico si la variable tiene un número almacenado
                elementos.append(float(numero_actual))  # Agrega el número actual a la lista de elementos
                numero_actual = ""  # Reinicia la construcción del numero_actual
            if caracter not in ["+", "-", "*", ":"]: # Verifico si el carácter no está presente en la lista
                print("Caracter no válido en la operación")  # Imprime un mensaje de error si el carácter no es un operador válido
                return  # Sale de la función si se encuentra un carácter no válido
            elementos.append(caracter)  # Agrega el operador a la lista de elementos

    if numero_actual: # Verifico si la variable tiene un número almacenado
        elementos.append(float(numero_actual))  # Agrega el último numero_actual a la lista de elementos

    resultado = elementos[0]  # Inicializa el resultado con el primer elemento de la lista
    i = 1
    while i < len(elementos):
        operador = elementos[i]  # Obtiene el operador actual
        numero = elementos[i + 1]  # Obtiene el número siguiente
        if operador == "+":
            resultado = resultado + numero  # Realiza la suma
        elif operador == "-":
            resultado = resultado - numero  # Realiza la resta
        elif operador == "*":
            resultado = resultado * numero  # Realiza la multiplicación
        elif operador == ":":
            if numero == 0:
                print("Error: No se puede dividir entre cero.")  # Imprime un mensaje de error si se intenta dividir entre cero
                return  # Sale de la función si se intenta dividir entre cero
            resultado = resultado / numero  # Realiza la división
        i = i + 2

    imprimir_resultado = input("Ingrese '=' para obtener el resultado: ")
    if imprimir_resultado == "=":
        print("El resultado final es:", resultado)  # Imprime el resultado final si se ingresó correctamente el símbolo "="
    
    else:
        print("Operación inválida.")  # Imprime un mensaje de error si no se ingresó correctamente el símbolo "="

```

## Calculadora de Fracciones

Primero Solicita al usuario que ingrese la operación que desea realizar en el formato "numerador/denominador operador numerador/denominador", por ejemplo, "2/3 + 1/4 : 1/2". Los elementos de la operación deben estar separados por espacios.


Luego, el programa separa los elementos de la operación en una lista llamada elementos. Los elementos pueden ser números enteros o fracciones representadas como cadenas de caracteres en el formato "numerador/denominador".


Recorre los elementos de la lista elementos y realiza las operaciones correspondientes. El numerador y denominador del resultado se actualizan en cada ciclo.

* Para la suma y resta, se calcula el común denominador y se realiza la operación.
* Para la multiplicación, se multiplica el numerador y denominador del resultado por los correspondientes del elemento actual.
* Para la división, se multiplica el numerador del resultado por el denominador del elemento actual y se multiplica el denominador del resultado por el numerador del elemento actual.

Finalmente, se solicita al usuario que ingrese "=" para obtener el resultado. Se muestra el resultado final en forma de fracción (si el denominador no es 1) o como un número entero (si el denominador es 1).

```python
def calculadora_fracciones():
    """
    Realiza operaciones de suma, resta, multiplicación y división de fracciones
    El programa se ejecuta hasta que el usuario elige la opción "4" para salir.
    """

    # Solicitar la operación al usuario.
    print("¡Bienvenido a la calculadora de fracciones!")
    operacion = input("Ingrese la operación que desea realizar (ejemplo: 2/3 + 1/4 : 1/2) : ")
    operacion = operacion.replace(" ", "")  # Elimina los espacios en blanco de la operación ingresada


    elementos = []  # Lista para almacenar los elementos de la operación
    numero_actual = ""  # Variable para construir el número actual
    for caracter in operacion:
        if caracter.isdigit() or caracter == "/": #. 
            numero_actual = numero_actual + caracter  # Agrega el carácter a la construcción del número actual si es un dígito o una barra de fracción
        else:
            if numero_actual:
                elementos.append(numero_actual)  # Agrega el número actual a la lista de elementos
                numero_actual = ""
            if caracter not in ["+", "-", "*", ":"]:
                print("Carácter no válido en la operación.")  # Imprime un mensaje de error si el carácter no es un operador válido
                return

            elementos.append(caracter)  # Agrega el operador a la lista de elementos

    if numero_actual:
        elementos.append(numero_actual)  # Agrega el último número actual a la lista de elementos

    resultado_num = 0  # Numerador del resultado
    resultado_den = 1  # Denominador del resultado
    i = 0
    while i < len(elementos):
        elemento_actual = elementos[i]  # Obtiene el elemento actual

        if "/" in elemento_actual:
            num, den = elemento_actual.split("/")  # Separa el numerador y el denominador de la fracción
            num = int(num)  # Convierte el numerador a entero
            den = int(den)  # Convierte el denominador a entero
        else:
            num = int(elemento_actual)  # Convierte el elemento a entero (si no es una fracción)
            den = 1  # El denominador es 1 para números enteros

        if i == 0: # Primer fracción
            resultado_num = num
            resultado_den = den
        else:
            operador = elementos[i - 1]  # Obtiene el operador anterior

            if operador == "+":
                den_comun = resultado_den * den
                resultado_num = resultado_num * den + num * resultado_den
                resultado_den = den_comun

            elif operador == "-":
                den_comun = resultado_den * den
                resultado_num = resultado_num * den - num * resultado_den
                resultado_den = den_comun

            elif operador == "*":
                # Multiplica fracciones
                resultado_num = resultado_num * num
                resultado_den = resultado_den * den

            elif operador == ":":
                if num == 0:
                    print("Error: No se puede dividir entre cero.")  # Imprime un mensaje de error si se intenta dividir entre cero
                    return
                # Divide fracciones
                resultado_num = resultado_num * den
                resultado_den = resultado_den * num

        i = i + 2
   
    siguiente_operacion = input("Ingrese '=' para obtener el resultado: ")
    if siguiente_operacion == "=":
        if resultado_den == 1:
            print("El resultado final es:", resultado_num)
        else:
            print("El resultado final es:", f"{resultado_num}/{resultado_den}") 
    else:
        print("Operación inválida.")
```

## Calculadora de Conversiones
 El usuario podrá elegir entre las siguientes opciones de conversión: 
 1. Decimal a Binario
 2. Decimal a Hexadecimal
 3. Decimal a Octal
 4. Salir
 
 Si eligio Salir volverá al menu principal, sino, deberá ingresar el numero decimal que desea convertir.
 Si el valor ingresado es válido, se almacena la información sobre el signo del numero y se realiza la conversión.

* Para la conversión a binario, se realiza una serie de divisiones enteras por 2 y se concatenan los residuos en orden inverso.
* Para la conversión a hexadecimal, se realizan divisiones enteras por 16 y se obtienen los dígitos hexadecimales correspondientes a los residuos, que se concatenan en orden inverso.
* Para la conversión a octal, se realizan divisiones enteras por 8 y se concatenan los residuos en orden inverso.

Después de realizar la conversión, se agrega el signo al resultado si el número original era negativo.

Luego, se solicita al usuario que ingrese "=" para mostrar el resultado. Si se ingresa "=", se muestra el resultado de la conversión y la función sale. Si se ingresa cualquier otro valor, se muestra un mensaje de error y la función sale.


```python
def calculadora_conversiones():
    """
    Convierte un número decimal a binario, hexadecimal u octal según la opción seleccionada.
    El programa se ejecuta en un bucle while hasta que el usuario elige la opción "4" para salir.
    """
    while True:
        # Mostrar el menú de opciones
        print("Calculadora de conversiones")
        print("Por favor, seleccione una opción:")
        print("1. Decimal a Binario")
        print("2. Decimal a Hexadecimal")
        print("3. Decimal a Octal")
        print("4. Salir")

        # Solicitar la opción al usuario
        opcion = input("Ingrese el número de la conversión que desea realizar: ")

        # Salir del programa si la opción es "4"
        if opcion == "4":
            return

        # Verificar la opción seleccionada
        elif opcion in ["1", "2", "3"]:
            try:
                # Solicitar el número al usuario y convertirlo a entero
                numero_decimal = int(input("Ingrese el número decimal que desea convertir: "))
            except:
                print("¡Error! El valor ingresado no es un número decimal válido.")
                return

            # Almacenar la información sobre el signo del número
            signo = ""
            if numero_decimal < 0:
                signo = "-"
                numero_decimal = abs(numero_decimal) # Función valor absoluto

            # Realizar la conversión según la opción seleccionada
            resultado = ""

            # Conversión a binario
            if opcion == "1":
                while numero_decimal > 0:
                    residuo = numero_decimal % 2  # Obtener el residuo de la división por 2
                    resultado = str(residuo) + resultado  # Concatenar el residuo al resultado
                    numero_decimal = numero_decimal // 2  # Realizar la división entera por 2

            # Conversión a hexadecimal
            elif opcion == "2":
                hex_digitos = "0123456789ABCDEF" # Cadena de caracteres hexadecimales
                while numero_decimal > 0:
                    residuo = numero_decimal % 16  # Obtener el residuo de la división por 16
                    resultado = hex_digitos[residuo] + resultado  # Obtener el dígito hexadecimal correspondiente al residuo y concatenarlo al resultado                 
                    numero_decimal = numero_decimal // 16  # Realizar la división entera por 16

            # Conversión a octal
            elif opcion == "3":
                while numero_decimal > 0:
                    residuo = numero_decimal % 8  # Obtener el residuo de la división por 8
                    resultado = str(residuo) + resultado  # Concatenar el residuo al resultado
                    numero_decimal = numero_decimal // 8  # Realizar la división entera por 8

            # Agregar el signo al resultado si el número original era negativo
            resultado = signo + resultado

            # Solicitar '=' al usuario para mostrar el resultado
            imprimir_resultado = input("Ingrese '=' para obtener el resultado: ")
            if imprimir_resultado == "=":
                # Mostrar el resultado de la conversión
                print("El resultado es:", resultado)
                return
            else:
                print("¡Error! La opción ingresada no es válida.")
                return

        else:
            print("¡Error! La opción ingresada no es válida.")
```

```python
while True:
    print("Bienvenido a la calculadora digital, Elija una función:")
    print("1. Clásica")
    print("2. Fracciones")
    print("3. Conversiones")
    print("4. Salir (off)")

    menu=input("Ingrese el número de la función que desea utilizar: ")
    if menu == "4":
        print ("Gracias por utilizar la calculadora digital")
        break
    elif menu == "1":
        calculadora_clasica()
    elif menu == "2":
        calculadora_fracciones()
    elif menu == "3":
        calculadora_conversiones()
    else:
        print (menu,": Es un caracter no válido")
```
